```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

path = "results/res.csv"

data = pd.read_csv(path)
print(data)
```

```python

"""
    Evaluation   Unbounded_1   Unbounded_2       Bounded  Bounded_log2
0  average_time(s)  3.489770e-02  1.431100e-02  2.536660e-02  2.208000e-02
1     average_calc  1.001898e+07  1.663938e+06  4.495320e+06  2.259770e+06
2   average_search  5.013708e+06  1.998000e+06  3.248763e+06  3.172141e+06
"""

# bar chart for average_time(s) for different algorithms
x = np.arange(4)
labels = data.columns[1:]
pltdata = data.values[0][1:]
plt.bar(x, pltdata)
plt.xticks(x, labels)
plt.ylabel("average_time(s)")
plt.title("average_time(s) for different algorithms")
plt.show()

```
